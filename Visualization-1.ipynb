```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read your Excel file into a DataFrame
df = pd.read_excel("C:/Users/tejas/Downloads/incidents-details-data.xlsx")

# Perform random sampling of 500 rows
sampled_df = df.sample(n=500, random_state=42)  # Select 500 random rows, using random_state for reproducibility

# Select specific unique values from the 'Incident' column
selected_incidents = ['Abuse', 'Assault', 'Vandalism', 'Robbery']

# Filter the DataFrame to include only rows with the selected unique values
filtered_df = sampled_df[sampled_df['Incident'].isin(selected_incidents)]

# Group by 'Perpetrator' and count occurrences of each 'Incident'
incident_perpetrator_counts = filtered_df.groupby(['Perpetrator', 'Incident']).size().unstack(fill_value=0)

# Grouped Bar Chart
incident_perpetrator_counts.plot(kind='bar', figsize=(12, 8))
plt.title('Incidents by Perpetrator')
plt.xlabel('Perpetrator')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Incident')
plt.tight_layout()
plt.show()

# Heatmap
plt.figure(figsize=(12, 8))
heatmap_data = filtered_df.groupby(['Perpetrator', 'Incident']).size().unstack(fill_value=0)
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='d')
plt.title('Incidents Heatmap by Perpetrator')
plt.xlabel('Incident')
plt.ylabel('Perpetrator')
plt.tight_layout()
plt.show()

# Pie Chart
incident_counts = filtered_df['Incident'].value_counts()
incident_counts.plot(kind='pie', figsize=(10, 8), autopct='%1.1f%%')
plt.title('Distribution of Incidents')
plt.ylabel('')
plt.tight_layout()
plt.show()

# Stacked Area Chart
stacked_df = filtered_df.groupby(['Perpetrator', 'Incident']).size().unstack(fill_value=0)

stacked_df.plot(kind='area', stacked=True, figsize=(12, 8))
plt.title('Incidents Trend by Perpetrator')
plt.xlabel('Perpetrator')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Incident')
plt.tight_layout()
plt.show()
```
